```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim

# 데이터셋 불러오기
df = pd.read_csv('전라북도_휴일약국_20170124..csv', encoding='euc-kr', nrows=3000)

# NaN 값 제거
df = df.dropna(subset=['주소'])

# 특정 지역명 입력
target_region = input("특정 지역명을 입력하세요: ")

# 특정 지역의 심야약국 정보 가져오기
target_pharmacies = df[df['주소'].str.contains(target_region)]

# 지오코더 초기화
geolocator = Nominatim(user_agent="myGeocoder")

# 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)  # 서울시 좌표, 원하는 위치로 변경해주세요

# 심야약국 위치 표시
for idx, row in target_pharmacies.iterrows():
    location = geolocator.geocode(row['주소'])
    if location is not None:
        folium.Marker([location.latitude, location.longitude], tooltip=row['약국명'], icon=folium.Icon(color='green')).add_to(m)

# HTML 파일로 저장
m.save('night_pharmacy_map.html')

# 지도 출력
m
```

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
import datetime

# 데이터셋 불러오기
df = pd.read_csv('전라북도_휴일약국_20170124..csv', encoding='euc-kr')

# NaN 값 제거
df = df.dropna(subset=['도로명주소'])

# 특정 지역명 입력
target_region = input("특정 지역명을 입력하세요(전라북도 내): ")

# 특정 지역의 심야약국 정보 가져오기
target_pharmacies = df[df['도로명주소'].str.contains(target_region)]

# 지오코더 초기화
geolocator = Nominatim(user_agent="myGeocoder")

# 지도 생성
m = folium.Map(location=[35.8347, 127.1171], zoom_start=10)  # 중심 좌표

# 심야약국 위치 표시
for idx, row in target_pharmacies.iterrows():
    location = geolocator.geocode(row['도로명주소'])
    if location is not None:
        # 운영시간 확인
        now = datetime.datetime.now()
        operating_time = row['운영시간']
        if operating_time and isinstance(operating_time, str):
            start_time = datetime.datetime.strptime(operating_time.split('~')[0], '%H:%M')
            if now.time() > start_time.time():
                tooltip_html = f"<strong>{row['약국명']}</strong><br>" \
                               f"도로명 주소: {row['도로명주소']}<br>" \
                               f"전화번호: {row['전화번호']}<br>" \
                               f"운영시간: {row['운영시간']}<br>" \
                               f"근무일: {row['근무일']}<br>" \
                               f"근무주기: {row['근무주기']}"
                folium.Marker([location.latitude, location.longitude], tooltip=tooltip_html, icon=folium.Icon(color='green')).add_to(m)

# HTML 파일로 저장
m.save('night_pharmacy_map.html')

# 지도 출력
m
```

```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
import datetime

# 데이터셋 불러오기
df = pd.read_csv('전라북도_휴일약국_20170124..csv', encoding='euc-kr')

# NaN 값 제거
df = df.dropna(subset=['도로명주소'])

# 특정 지역명 입력
target_region = input("특정 지역명을 입력하세요(전라북도 내): ")

# 특정 지역의 심야약국 정보 가져오기
target_pharmacies = df[df['도로명주소'].str.contains(target_region)]

# 지오코더 초기화
geolocator = Nominatim(user_agent="myGeocoder")

# 지도 생성
m = folium.Map(location=[35.8347, 127.1171], zoom_start=10)  # 중심 좌표

# 심야약국 위치 표시
for idx, row in target_pharmacies.iterrows():
    location = geolocator.geocode(row['도로명주소'])
    if location is not None:
        # 운영시간 확인
        now = datetime.datetime.now()
        operating_time = row['운영시간']
        if operating_time and isinstance(operating_time, str):
            start_time_str, end_time_str = operating_time.split('~')
            start_time = datetime.datetime.strptime(start_time_str.strip(), '%H:%M')
            end_time = datetime.datetime.strptime(end_time_str.strip(), '%H:%M')
            if end_time.time() >= datetime.time(18, 0):  # 운영시간이 18시 이후인지 확인
                tooltip_html = f"<strong>{row['약국명']}</strong><br>" \
                               f"도로명 주소: {row['도로명주소']}<br>" \
                               f"전화번호: {row['전화번호']}<br>" \
                               f"운영시간: {row['운영시간']}<br>" \
                               f"근무일: {row['근무일']}<br>" \
                               f"근무주기: {row['근무주기']}"
                folium.Marker([location.latitude, location.longitude], tooltip=tooltip_html, icon=folium.Icon(color='red', icon='ok-circle')).add_to(m)

# HTML 파일로 저장
m.save('night_pharmacy_map2.html')

# 지도 출력
m
```
