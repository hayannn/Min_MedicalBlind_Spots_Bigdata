```python
#인천시 미추홀구 동물약국 현황 오픈 API 사용
import folium
import requests
import xml.etree.ElementTree as ET

API_KEY = 'e4isidJd3OhFK2lR1lOxhd8X2Vg4sxZZoW2XliIqlK+wcO15Yj3aOH/osxWF9dGy2zHdpOkquNrXBb3Pe5X8qg=='

# API 호출을 위한 URL
url = 'http://apis.data.go.kr/3510500/animal_pharmacy/getList'

# 필요한 파라미터 설정
params = {
    'serviceKey': API_KEY,
    'page': '1',
    'numOfRows' : '10',
    'type' : 'XML'
}

# API 호출
response = requests.get(url, params=params)

try:
    root = ET.fromstring(response.text)
    items = root.find('.//body/items')
except ET.ParseError as e:
    print("API 호출에 실패했습니다:", e)
    items = None

if items:
    # 지도 초기 설정
    map = folium.Map(location=[37.4531, 126.6647], zoom_start=13)

    # 약국 데이터 처리
    for item in items:
        name = item.find('fc_nm').text
        rep_persion = item.find('rep_person').text
        tel_no = item.find('tel_no').text
        lat = float(item.find('lat').text)
        lng = float(item.find('lot').text)
        addr = item.find('road_nm_addr').text
        
        tooltip_html = f"<strong>{name}</strong><br>" \
                       f"도로명 주소 : {addr}<br>" \
                        f"대표자 : {rep_persion}<br>" \
                       f"전화번호: {tel_no}"

        # 마커 추가
        folium.Marker([lat, lng], tooltip=tooltip_html,icon=folium.Icon(color='red')).add_to(map)

        
    # 지도 저장 및 출력
    map.save("animal_pharmacies_map.html")
    print("지도가 생성되었습니다. 'animal_pharmacies_map.html' 파일을 확인해주세요.")
    
map
```
